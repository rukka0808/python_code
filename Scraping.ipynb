```python
import requests
url = "http://www.kric.go.kr/jsp/industry/rss/citystapassList.jsp?q_org_cd=A010010021&q_fdate=2021"
```

```python

html_text = requests.get(url).text
```

```python
print(html_text)
```

```python
from bs4 import BeautifulSoup
```

```python
pip install bs4
```

```python
soup = BeautifulSoup(html_text, "html.parser")
```

```python
print(soup)
```

```python
tab = soup.find("table",{"class": "listtbl_c100"})
```

```python
print(tab)
```

```python
trs = tab.find('tbody').find_all('tr')
```

```python
trs[1]
```

```python
tdcols = trs[1].find("td",{"class":"tdcol"})
```

```python
print(tdcols)
```

```python
tds = trs[1].find_all('td')
print(tds)
```

```python
tds[0]
```

```python
tds[0].text
```

```python
tds[2].text
```

위 내용을 리스트를 만들어 반복문으로 저장

```python
stationpassengers = []
for tr in trs[1:]:
  dic = {}
  tds = tr.find_all('td')
  dic['station'] = tds[0].text
  dic['ride'] = tds[2]. text
  dic['alight'] = tds[3].text
  stationpassengers.append(dic)
```

```python
json = response.json()
```

```python
cars = json['realUnitTrtmVO']
records = []
for car in cars:
  dic ={}
  dic['date'] = car['stdDate']
  dic['time'] = car['stdTime']
  dic['type'] = car['tcsCarTypeDivName']
  records.append(dic)
records
```
